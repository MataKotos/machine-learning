На каком континенте в среднем потребляется больше всего пива?

```python
df_clean = df[df['continent'] != 0]

df_clean.groupby('continent')['beer_servings'].mean().sort_values(ascending=False)
```

Напечатайте статистику потребления вина (wine_servings) для каждого континента

```python
df['continent'] = df['continent'].replace(0, np.nan)
df = df.dropna(subset=['continent'])
df.groupby('continent')['wine_servings'].describe()


```

Напечатайте среднее потребление каждой категории вин на каждом
континенте. возможно имелись ввиду категории алкоголя, тк их несколько, а категорий вин нет...

```python
df.groupby('continent')[
    ['beer_servings', 'spirit_servings', 'wine_servings']
].mean()
```

Напечатайте медианное потребление каждой категории алкоголя на каждом континенте

```python
df.groupby('continent')[
    ['beer_servings', 'spirit_servings', 'wine_servings']
].median()
```

Напечатайте среднее, максимальное и минимальное потребление
спиртных напитков для каждого континента

```python
df.groupby('continent')['spirit_servings'].agg(['mean', 'max', 'min'])
```

Какая страна в среднем потребляет больше всего спирта?

```python
df.loc[df['spirit_servings'].idxmax(), 'country']
```

Выведите на экран список «безалкогольных» стран

Вычислите для столбца total_litres_of_pure_alcohol среднее значение,
моду, медиану, дисперсию и среднеквадратичное отклонение

```python
column = df['total_litres_of_pure_alcohol']

mean_value = column.mean()
mode_value = column.mode()
median_value = column.median()
variance_value = column.var()
std_value = column.std()

mean_value, mode_value, median_value, variance_value, std_value
```
